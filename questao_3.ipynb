```python
import pandas as pd
from transformers import pipeline
```

```python

data = pd.read_csv('dataset/articles.csv')
data
```

```python
filteredData = data[(data['category'] == 'mercado') & (data['date'] >= '2015-01-01') & (data['date'] <= '2015-03-31')]
filteredData
```

```python
classifier = pipeline("ner", model="monilouise/ner_pt_br", grouped_entities=True)

entities = []

for text in filteredData['text']:
    c = (classifier(text))
    for e in c:
        if(e['entity_group'] == 'ORG')  and not e['word'].startswith("##")  and len(e['word']) > 1:
            entities.append(e['word']);
```

```python
organizacoes = {}

for org in entities:
    if org in organizacoes:
        organizacoes[org] += 1
    else:
        organizacoes[org] = 1

organizacoes = sorted(organizacoes.items(), key=lambda item: item[1], reverse=True)

organizacoes
```
